```python
print("Welcome to the Library Management System \n\n\n Please login to access the system")
def login():
    credentials = {"admin":"admin123","librarian" :"lib123"}
    attempts = 3
    while attempts >0:
        username = input("Enter username:")
        password = input("Ener password:")
        if username in credentials and credentials[username] == password:
            print("Login Succesful! Exploare the Library!")
            return True
        else:
            attempts -= 1
            print("Invalid credentials", attempts, "left")
        return False
    print("Too many failed attempt Please try after sometime")

def main_menu():
    if not login():
        print("Please login")
    while True:
        print("\n========= LIBRARY MENU =========")
        print("1. View Books")
        print("2. Add a Book")
        print("3. Issue a Book")
        print("4. Return a Book")
        print("5. View Issued Books")
        print("6. Exit")

        choice = input("Enter your option(1-6):").strip()
        if choice == "1":
            view_books()
        elif choice == "2":
            add_book()
        elif choice == "3":
            issue_book()
        elif choice == "4":
            return_book()
        elif choice == "5":
            view_issued_book()
        elif choice == "6":
            print("Thanks for using Library System, Good bye!")
            break
        else:
            print("Pleas enter a Valid input from numbers 1 to 6")
library = {
    "Python Basics": {"author": "David Amos", "availability": True, "issued_to": None},
    "Data Science": {"author": "John D. Kelleher", "availability": False, "issued_to": "Rahul"},
    "Machine Learning" : {"author": "Kevin", "availability" : True, "issued_to": None},
    "Deep Learning with Python":{"author" :"Francois Chollet","availability" : False, "issued_to":"Sanvi"}
}

# Step 2: Display all books
def view_books():
    print(" "*30 , "Library Books List"," "*30)
    print(f"{'Title':<25}{'Author':<25}{'Status':<15}{'Issued To'}")
    print("-" * 90)
    for title, info in library.items():
        status = "Available" if info["availability"] else "Issued"
        issued_to = info["issued_to"] if info["issued_to"] else "-"
        print(f"{title:<25}{info['author']:<25}{status:<15}{issued_to}")
    print("-" * 70)

# add new book
def add_book():
    title = input("\nEnter book title: ").strip()
    author = input("Enter author name: ").strip()
    if title in library:
        print("Book already exists in the library.")
    else:
        library[title] = {"author": author, "availability": True, "issued_to": None}
        print(f"{title} by {author} added successfully!")
# Issue a book
def issue_book():
    book_name = input("Enter the book name you want to issue:")
    if book_name in library:
        if library[book_name]["availability"]:
            person = input("Enter the person name issuing the book:").strip()
            library[book_name]['availability']= False
            library[book_name]['issued_to'] = person
            print(book_name," book issued to ", person)
        else:
            print(book_name, "is already issued to", library[book_name]["issued_to"])
    else:
        print(book_name,"book not found in the library")     

# View Issued books List
def view_issued_book():
    print("\n" + " " * 25 + "---- Issued Books List ----" + " " * 25)
    print("-" * 60)
    print(f"{'Title':<30}{'Issued To':<25}")
    print("-" * 60)

    found = False
    for title, info in library.items():
        if not info["availability"]:      
            found = True
            print(f"{title:<30}{info['issued_to']:<25}")

    if not found:
        print("No books are currently issued.")

    print("-" * 60)

def return_book():
    title = input("Enter the book title to return: ").strip()
    if title in library:
        if not library[title]["availability"]:
            library[title]["availability"] = True
            person = library[title]["issued_to"]
            library[title]["issued_to"] = None
            print(f"{title} has been returned by {person}.")
        else:
            print("That book was not issued.")
    else:
        print("Book not found in the library.")


```

```python
main_menu()
```
